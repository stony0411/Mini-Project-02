```python
import os
import os.path
import xml.dom.minidom
from os import listdir
from os.path import isfile, join
```

```python
# Renumber all imgs in this folder, imcluding sub folders
path = r"C:\Users\stony\Downloads\Dataset_UAV123_10fps\UAV123_10fps\data_seq\UAV123_10fps"
count = 1
for dir in os.listdir(path):
    for img in os.listdir(os.path.join(path, dir)):
        img_path = os.path.join(os.path.join(path, dir, img))
        new_path = os.path.join(os.path.join(path, dir, str(count).zfill(6)+".jpg"))
        os.rename(img_path, new_path)
        count+=1
```

```python
# Change the bbox in the xml files
path=r'C:\Users\stony\Downloads\Dataset_UAV123_10fps\UAV123_10fps\data_seq\UAV123_10fps\wakeboard10'
txt=open(r"C:\Users\stony\Downloads\Dataset_UAV123_10fps\UAV123_10fps\anno\UAV123_10fps\wakeboard10.txt", "r")
files=os.listdir(path)
for xmlFile in files: 
    filename=os.path.splitext(xmlFile)[0]   
    filetype=os.path.splitext(xmlFile)[1]
    if filetype == '.xml' :
        print (xmlFile)
        dom=xml.dom.minidom.parse(os.path.join(path,xmlFile))
        root=dom.documentElement
        xmin=root.getElementsByTagName('xmin')
        ymin=root.getElementsByTagName('ymin')
        xmax=root.getElementsByTagName('xmax')
        ymax=root.getElementsByTagName('ymax')
        content = txt.readline()
        content_list = content.split(",")
    
        if content_list[0].strip()=='NaN':
            os.remove(os.path.join(path,xmlFile))
        else :
            xmin[0].firstChild.data=int(content_list[0].strip())
            ymin[0].firstChild.data=int(content_list[1].strip())
            xmax[0].firstChild.data=int(content_list[0].strip())+int(content_list[2].strip())
            ymax[0].firstChild.data=int(content_list[1].strip())+int(content_list[3].strip())

            #SAVE
            with open(os.path.join(path,xmlFile),'w') as fh:
                dom.writexml(fh)

txt.close()          
```
