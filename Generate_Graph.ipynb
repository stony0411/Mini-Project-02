```python
import os
import os.path
import re
import numpy as np
import matplotlib.pyplot as plt
txt=open(r"C:\Users\stony\Downloads\loss.txt", "r")

train_loss_total=[]
train_loss_giou=[]
train_loss_l1=[]
train_iou=[]

val_loss_total=[]
val_loss_giou=[]
val_loss_l1=[]
val_iou=[]

for line in txt:
    content_list = re.split(",|: |] | / ",line)
    #if not int(content_list[1])%2 :
    if int(content_list[2])==int(content_list[3]):
        print(content_list)
        if content_list[0]=='[train':
            train_loss_total.append(float(content_list[7]))
            train_loss_giou.append(float(content_list[9]))
            train_loss_l1.append(float(content_list[11]))
            train_iou.append(float(content_list[13]))
        elif content_list[0]=='[val':
            val_loss_total.append(float(content_list[7]))
            val_loss_giou.append(float(content_list[9]))
            val_loss_l1.append(float(content_list[11]))
            val_iou.append(float(content_list[13]))
```

```python
print(train_loss_total)
print(train_loss_giou)
print(train_loss_l1)
print(train_iou)
```

```python
print(val_loss_total)
print(val_loss_giou)
print(val_loss_l1)
print(val_iou)
```

```python
epoch = np.arange(0,50,1)
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.plot(epoch, train_loss_total)
#plt.plot(epoch, train_loss_giou,label='train_loss/giou')
#plt.legend(loc='upper right')
plt.title('loss/total vs. epoch')
plt.ylabel('loss/total')
plt.xlabel('epoch')
plt.subplot(2, 2, 2)
plt.plot(epoch, train_loss_giou)
#plt.plot(epoch, train_iou,label='IoU')
#plt.legend(loc='right')
plt.title("loss/giou vs. epoch")
plt.ylabel('loss/giou')
plt.xlabel('epoch')
plt.subplot(2, 2, 3)
plt.plot(epoch, train_loss_l1)
#plt.plot(epoch, val_loss_giou,label='val_loss/l1')
#plt.legend(loc='right')
plt.title('loss/l1 vs. epoches')
plt.ylabel('loss/l1')
plt.xlabel('epoch')
plt.subplot(2, 2, 4)
plt.plot(epoch, train_iou)
#plt.plot(epoch, val_loss_giou,label='val_IoU')
#plt.legend(loc='right')
plt.title('IoU vs. epoches')
plt.ylabel('IoU')
plt.xlabel('epoch')
plt.subplots_adjust(
                wspace=0.3, 
                hspace=0.3)
```
